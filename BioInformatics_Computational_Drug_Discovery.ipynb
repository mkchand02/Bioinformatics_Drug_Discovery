```python
#import libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search for coronavirus
target = new_client.target
target_query = target.search('coronavirus')#'acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
targets.columns
```

###Select and retrieve bioactivity data for SARS coronavirus 3C-like proteinase

its the seventh entry in the ***targets*** variable

```python
selected_target = targets.target_chembl_id[6]
selected_target
```

```python
selected_target_name = targets.pref_name[6]
selected_target_std_activity_type = "IC50"
selected_target_name = selected_target_name.replace("-","_")
selected_target_name = selected_target_name.replace(" ","_")
selected_target_name, selected_target_std_activity_type
```

```python
# Here, we will retrieve only bioactivity data that are reported as pChEMBL values
#   that are reported in IC50 values in nM (nano molar) unit.
activity = new_client.activity
res = activity.filter(target_chembl_id = selected_target).filter(standard_type = selected_target_std_activity_type)
```

```python
#create a dataframe of above mentioned bioactivity data
df = pd.DataFrame.from_dict(res)
df.columns
```

```python
df.head(5)
#lower the standard value (inhibitory concentration) , higher the potency
```

## Data pre-processing of the bioactivity data
###Classify compounds as Active, inactive and intermediate

* Compounds having bioactivity IC50 value less than 1000 nM are active
* Compounds having bioactivity IC50 value greater than 10000 nM are inactive
* Compounds having bioactivity IC50 value between 1000 nM and 10000 nM are intermediate


#####Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame

```python
bioactivity_threshold = []
for i in df2.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

###Combine the 3 columns (molecule_chembl_id,canonical_smiles,standard_value) and bioactivity_class into a DataFrame
The molecules referred here are compounds/drugs that produces effects on the target protein

```python


selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2[selection]
df3.to_csv('bioactivity_preprocessed_data.csv', index = False)
df3
```
